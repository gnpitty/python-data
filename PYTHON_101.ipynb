## 3. Operadores ‚ûï‚ûñ‚úñÔ∏è‚ûó

```python
a = 10
b = 3

print("Operadores Aritm√©ticos:")
print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
print(f"{a} * {b} = {a * b}")
print(f"{a} / {b} = {a / b}")
print(f"{a} // {b} = {a // b}")  # Divisi√≥n entera
print(f"{a} % {b} = {a % b}")    # M√≥dulo
print(f"{a} ** {b} = {a ** b}")  # Potencia

# Operadores de comparaci√≥n
print("\nOperadores de Comparaci√≥n:")
print(f"{a} > {b}: {a > b}")
print(f"{a} == {b}: {a == b}")
print(f"{a} != {b}: {a != b}")

# Operadores l√≥gicos
print("\nOperadores L√≥gicos:")
print(f"True and False: {True and False}")
print(f"True or False: {True or False}")
print(f"not True: {not True}")
```

### Condicionales (if, elif, else)

```python
edad = 18

if edad < 13:
    categoria = "Ni√±o"
elif edad < 20:
    categoria = "Adolescente"
elif edad < 60:
    categoria = "Adulto"
else:
    categoria = "Adulto mayor"

print(f"Con {edad} a√±os, eres: {categoria}")
```

```python
# Ejemplo con m√∫ltiples condiciones
temperatura = 25
lluvia = False

if temperatura > 20 and not lluvia:
         print("¬°Perfecto para salir!")
elif temperatura > 20 and lluvia:
    print("Hace calor pero llueve, lleva paraguas")
else:
    print("Mejor quedarse en casa")
    print("demo")
```

## 5. Funciones üîß

```python
# Funci√≥n b√°sica
def saludar(nombre):
    """Funci√≥n que saluda a una persona"""
    return f"¬°Hola, {nombre}!"

# Funci√≥n con par√°metros por defecto
def presentarse(nombre, edad=25, ciudad="Madrid"):
    """Funci√≥n con par√°metros por defecto"""
    return f"Me llamo {nombre}, tengo {edad} a√±os y vivo en {ciudad}"

# Funci√≥n que retorna m√∫ltiples valores
def operaciones_basicas(a, b):
    """Realiza operaciones b√°sicas entre dos n√∫meros"""
    suma = a + b
    resta = a - b
    multiplicacion = a * b
    return suma, resta, multiplicacion

# Probando las funciones
print(saludar("Mar√≠a"))
print(presentarse("Carlos"))
print(presentarse("Ana", 30, "Barcelona"))

resultado_suma, resultado_resta, resultado_mult = operaciones_basicas(10, 5)
print(f"10 + 5 = {resultado_suma}")
print(f"10 - 5 = {resultado_resta}")
print(f"10 * 5 = {resultado_mult}")
```

## 6. Estructuras de Datos üì¶

### 6.1 LISTAS üìù
Las listas son colecciones ordenadas y mutables (se pueden modificar)

```python
 #Crear listas
numeros = [1, 2, 3, 4, 5]
colores = ["rojo", "verde", "azul"]
mixta = [1, "texto", 3.14, True]

print("Listas creadas:")
print(f"N√∫meros: {numeros}")
print(f"Colores: {colores}")
print(f"Lista mixta: {mixta}")

# Acceder a elementos (indexing)
print(f"\nPrimer color: {colores[0]}")
print(f"√öltimo color: {colores[-1]}")
print(f"Segundo n√∫mero: {numeros[1]}")
```

```python
# Slicing (rebanadas)
print(f"\nPrimeros 3 n√∫meros: {numeros[:3]}")
print(f"√öltimos 2 n√∫meros: {numeros[-2:]}")
print(f"N√∫meros del 1 al 3: {numeros[1:4]}")

# M√©todos de listas
print("\n--- M√©todos de listas ---")
frutas = ["manzana", "banana"]
print(f"Lista original: {frutas}")

```

```python
# Agregar elementos
frutas.append("naranja")
print(f"Despu√©s de append: {frutas}")

frutas.insert(1, "pera")
print(f"Despu√©s de insert: {frutas}")

# Eliminar elementos
frutas.remove("banana")
print(f"Despu√©s de remove: {frutas}")

ultimo = frutas.pop()
print(f"Despu√©s de pop: {frutas}")
print(f"Elemento eliminado: {ultimo}")

# Longitud y b√∫squeda
print(f"\nLongitud de la lista: {len(frutas)}")
print(f"¬ø'manzana' est√° en la lista? {'manzana' in frutas}")
```

### 6.2 TUPLAS üîí
Las tuplas son colecciones ordenadas e inmutables (no se pueden modificar)

```python
# Crear tuplas
coordenadas = (10, 20)
persona = ("Juan", 25, "Ingeniero")
colores_tupla = ("rojo", "verde", "azul")

print("Tuplas creadas:")
print(f"Coordenadas: {coordenadas}")
print(f"Persona: {persona}")
print(f"Colores: {colores_tupla}")

# Acceder a elementos (igual que las listas)
print(f"\nCoordenada X: {coordenadas[0]}")
print(f"Nombre: {persona[0]}")
print(f"Edad: {persona[1]}")

# Desempaquetado de tuplas
x, y = coordenadas
nombre, edad, profesion = persona

print(f"\nDesempaquetado:")
print(f"X={x}, Y={y}")
print(f"Nombre: {nombre}, Edad: {edad}, Profesi√≥n: {profesion}")

# M√©todos de tuplas (limitados porque son inmutables)
print(f"\n--- M√©todos de tuplas ---")
numeros_tupla = (1, 2, 3, 2, 4, 2)
print(f"Tupla: {numeros_tupla}")
print(f"Contar '2': {numeros_tupla.count(2)}")
print(f"√çndice de '3': {numeros_tupla.index(3)}")
print(f"Longitud: {len(numeros_tupla)}")

# ¬øCu√°ndo usar tuplas?
print("\nüí° Usa tuplas cuando:")
print("- Los datos no cambiar√°n")
print("- Necesites coordenadas (x, y)")
print("- Quieras retornar m√∫ltiples valores de una funci√≥n")
```

### 6.3 DICCIONARIOS üóùÔ∏è
Los diccionarios almacenan pares clave-valor

```python
# Crear diccionarios
estudiante = {
    "nombre": "Mar√≠a",
    "edad": 22,
    "carrera": "Inform√°tica",
    "notas": [8.5, 9.0, 7.5]
}

# Diccionario vac√≠o
inventario = {}

print("Diccionario creado:")
print(estudiante)

# Acceder a valores
print(f"\nNombre: {estudiante['nombre']}")
print(f"Edad: {estudiante['edad']}")
print(f"Primera nota: {estudiante['notas'][0]}")

# M√©todo get() (m√°s seguro)
print(f"Carrera: {estudiante.get('carrera')}")
print(f"Email: {estudiante.get('email', 'No disponible')}")

# Agregar y modificar elementos
estudiante['email'] = 'maria@email.com'
estudiante['edad'] = 23
print(f"\nDiccionario actualizado:")
print(estudiante)

# M√©todos importantes
print(f"\n--- M√©todos de diccionarios ---")
print(f"Claves: {list(estudiante.keys())}")
print(f"Valores: {list(estudiante.values())}")
print(f"Pares clave-valor: {list(estudiante.items())}")

# Iterar sobre diccionarios
print("\nIterando sobre el diccionario:")
for clave, valor in estudiante.items():
    print(f"{clave}: {valor}")

for clave in estudiante.keys():
    print(f"{clave}: {estudiante[clave]}")

# Eliminar elementos
email = estudiante.pop('email')
print(f"\nEmail eliminado: {email}")
print(f"Diccionario despu√©s del pop: {estudiante}")

# Diccionario de inventario (ejemplo pr√°ctico)
print("\n--- Ejemplo pr√°ctico: Inventario ---")
inventario = {
    "manzanas": 50,
    "bananas": 30,
    "naranjas": 25
}

print("Inventario inicial:", inventario)

# Vender frutas
inventario["manzanas"] -= 10
inventario["bananas"] -= 5

# Agregar nueva fruta
inventario["peras"] = 20

print("Inventario actualizado:", inventario)
```
